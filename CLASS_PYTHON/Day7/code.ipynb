```python
import requests

url = "http://ekantipur.com"

response = requests.get(url)
if response.status_code == 200:
    data = response.text
    soup = BeautifulSoup(data , "html.parser")
    # print(soup.prettify())
    
else:
    print(f"Response {response.status_code}")
```

```python
import pandas as pd
data = soup.find_all("h2")

final_data = []
for val in data:
    # print(val.text)
    text_data = val.text
    link_data = val.a
    
    if link_data:
        link = link_data.get("href")
    # print(text_data , link)
    
    if len(text_data) > 20:
        final_data.append({
            "headline":text_data , 
            "link":link
        })

df = pd.DataFrame(final_data)
df.to_csv("news.csv" , index=False)
```

```python
import pandas as pd
import csv
import random
import names
import numpy as np 

aList = []

for i in range (1 , 100001):
    aList.append({
        "id":i,
        "name":names.get_full_name(),
        "salary":10000*i
    })

df = pd.DataFrame(aList)
df.to_csv("largeData.csv",index=False)
```

```python
file_name = "largeData.csv"
output_file = "output_data.csv"

first_chunk = True

for chunk in pd.read_csv(file_name, chunksize=10000):
    chunk["Increased_Salary"] = chunk["salary"] * 1.10
    chunk.to_csv(output_file, mode="w" if first_chunk else "a", index=False)
```
