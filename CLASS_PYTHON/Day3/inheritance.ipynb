```python
# Classical Example of Simple Inheritance

class Animal:
    def __init__(self , name):
        self.name = name
    
    def speak(self):
        print(f"{self.name} is speaking")
    
    
    
    # Derived Class 
# base class ma constructor xa vaney derived class ma nabanaye pani hunxa.

class Dog(Animal): # simple inheritance.
    def __init__(self, name , breed):
        super().__init__(name)
        self.breed = breed
    
    
    def speak(self):
        super().speak()
        print(f"{self.name} of {self.breed} is Barking")  # this is an example of function overriding
    

# base class ko same naam ko function derived class ma pani same to same name ko banaunu lai nei func overrriding vaninxa.
# super() ->  aafu vanda mathi ko function call garxa.


# dog ko chai init call vayo. and then super() function call vayo.
# to cancel override or to call the parent one we use the super() function 

d1 = Dog("Lucifer" , "Golden Retriever")
d1.speak()
```

```python
# Multiple Inheritance

class A :
    def show(self):
        print("A")
    

class B: 
    def show(self):
        print("B")
        
class C(B , A):
    # Jun lai agaadi use garinxa teskei nei function call hunxa
    """
    Under the hood python le list create garxa.
    """
    pass

c1 = C()
c1.show()
```

```python
# during inheritance order of the class created is shown by this mro() function
C.mro()
```

```python
class DataList:
    def __init__(self , list_1):
        self.list_1 = list_1
        
    
    def mean(self):
        sum = 0
        for i in self.list_1:
            sum+=i
        length = len(self.list_1)
        avg = sum/length
        return f"The mean is {avg}"
    
# Alternate method
# return sum(self.list) / len(self.list)
    

l1 = DataList([1,2,3,4,5])
l1.mean()
```

```python
# Exercise 1.5

# Exception Handling (try , except)
while True:
    try:
        n1 = input("Enter first number: ")
        n2 = input("Enter second number: ")
        
        if n1 == "q"  and n2 == 'q':
            break
        result = int(n1) / int(n2)
        print(f"The value is :{result} ")
    except ZeroDivisionError:
        print(f"Enter the valid number which shouldn't contain denom as 0 ")
    except ValueError:
        print(f"laude enter number only.")
    finally:
        print("This always runs")
```
