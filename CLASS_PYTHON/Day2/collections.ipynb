```python
from collections import Counter

alist = [1,2,1,"Ram" , "Shyam" , "Hari" , "Ram"]
count = Counter(alist)


print(count[4])
count[6] = 23
print(count[6])
print(type(count))
```

```python
for keys in count.keys():
    print(keys)
for values in count.values():
    print(values)

for items in count.items():
    print(items)
```

```python
#Default dictionary. 

"""
First let's learn a disadvantages of dictionary
"""

# aDict = {}
# aDict['a'] = aDict['a'] + 1

# this means there is no default values.

# To solve this 


from collections import defaultdict

bDict = defaultdict(int)
bDict['a'] = bDict['a'] + 1


strDict = defaultdict(str)
strDict['name'] = strDict['name'] + "Himal"
print(strDict)
print(bDict)
```

```python
#Double Ended Queue   -> queue but duitei mukh baata haalna milxa duitei bata nikaalna milxa.

from collections import deque

queue = deque([1,5,7])
print(queue)
queue.append(112)
print(queue)
queue.appendleft(123)
print(queue)
queue.pop()
print(queue)
queue.popleft()
print(queue)
```

```python
while queue:
    val = queue.pop()
    print(val)
```
