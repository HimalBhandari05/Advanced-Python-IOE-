```python
import numpy as np 

x = np.array([1900, 1000 , 2333 , 1233])

transformed = np.log1p(x)
print(transformed)
```

```python
import numpy as np 
from sklearn.preprocessing import PowerTransformer

x = np.array([1900, 1000 , 2333 , 1233])
pt=PowerTransformer(method='yeo-johnson')
transformed = pt.fit_transform(x.reshape(-1,1))
print(transformed)
```
