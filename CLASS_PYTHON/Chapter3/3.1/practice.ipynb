```python
import requests

url = "https://quotes.toscrape.com/"

response = requests.get(url)

if response.status_code == 200:
    data = response.text
    soup = BeautifulSoup(data, "html.parser")
    print(soup.prettify())

else:
    print(f"Response {response.status_codes }")
```

```python
import pandas as pd
import pprint

quotes_div = soup.find_all("div", class_="quote")

quotes = []
author = []



for quote_div in quotes_div:
    quote_text = quote_div.find("span", class_="text").get_text(strip=True)
    quote_author = quote_div.find("small", class_="author").get_text(strip=True)

    quotes.append(quote_text)
    author.append(quote_author)


print(pprint.pprint(quotes))

print("\n\n\n next \n")

print(pprint.pprint(author))

final_data = []

final_data.append(author)
final_data.append(quotes)
```

```python
import pandas as pd


df = pd.DataFrame(final_data)
df.to_csv("quotes.csv" , index=False)
```
