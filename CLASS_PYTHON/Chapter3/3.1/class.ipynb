```python
import pandas as pd 
import matplotlib.pyplot as plt

data ={
    "Name":["Ram" , "Shyam" , "Hari"],
    "Salary":["$2000" ,"$3000" , "$5000"]
}

df = pd.DataFrame(data)
df.to_csv("file.csv" , index=False)
print(df.head())
```

```python
df_A = pd.DataFrame({
    "Roll":[1,2,3],
    "Name":["Ram" , "Shyam" , "Hari"]
})

df_B = pd.DataFrame({
    "Roll":[1,2,5],
    "Percentage":[80 ,23,90]
})

inner_merge = pd.merge(df_A  , df_B , on="Roll", how="inner")
print(inner_merge)
```

```python
outer_merge = pd.merge(df_A  , df_B , on="Roll", how="outer")
print(outer_merge)
```
