```python
df_A = pd.DataFrame({
    "Roll":[1,2,3],
    "Name":["Ram" , "Shyam" , "Hari"]
})

df_B = pd.DataFrame({
    "Roll":[1,2,5],
    "Percentage":[80 ,23,90]
})

inner_merge = pd.merge(df_A  , df_B , on="Roll", how="inner") #intersection
print(inner_merge)
```

```python
outer_merge = pd.merge(df_A  , df_B , on="Roll", how="outer") # union
print(outer_merge)
```

```python
# left-> a ma matra vako right -> b ma matra vako 

left_merge = pd.merge(df_A , df_B , on="Roll" , how="left")
print(left_merge)
```

```python
import pandas as pd

df_a = pd.DataFrame(
    {"Temperature": [10, 123, 34, 54]}, index=["Day1", "Day2", "Day3", "Day4"]
)

df_b = pd.DataFrame(
    {"Humidity": [100, 1223, 344, 564]}, index=["Day1", "Day3", "Day6", "Day8"]
)

joined_df = df_a.join(df_b, how="outer")
print(joined_df)
```

```python
# melt vaneko wide lai long format ma laaney. 

wide_df = pd.DataFrame(
    {
        "Name":["Himal","Sajina","Jina","Jitan"],
        "Age":[20 , 20 , 17 , 44],
        "Address":["Jhapa","Jhapa","Morang","Sitapuri"]
    }
)


print(wide_df)
```

```python
# to change the wide to long format we have. 

long_df = pd.melt(
    frame=wide_df,
    id_vars="Name",
 )

print(long_df)
```

```python
import pandas as pd

df_wide = pd.DataFrame({
    "student": ["Ram", "Sita"],
    "Math": [85, 88],
    "Science": [90, 92],
    "English": [78, 81]
})

print(df_wide)


long_df = pd.melt(
    df_wide , 
    id_vars="student",
    value_vars=["Math","Science","English"],
    var_name="subject",
    value_name="score"
)
```

```python
print(long_df)
```

```python
stacked_df = long_df.stack()
print(stacked_df)
```

```python
unstacked_df = stacked_df.unstack()
print(unstacked_df)
```

```python
pivot_tbl = long_df.pivot_table(
    index="student",
    columns="subject",
    values="score",
    aggfunc= "mean"   
)

print(pivot_tbl)
```
