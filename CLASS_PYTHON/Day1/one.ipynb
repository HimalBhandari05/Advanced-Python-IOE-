```python
#args and kwargs 
#agrs - takes tuple as an argument whereas Kwargs - takes dictionary as an argument 

def average(*args):
    sum = 0
    length = len(args)
    
    for i in args:
        sum += i 
    average = sum / length
    print(average)

average(1,34,53,2)
```

```python
# wap to find average using kwargs

def average(**kwargs):
    sum = 0 
    for i in kwargs.keys():
        sum += kwargs[i]
    print(sum)
    
average(a = 10 , b = 234)
```

```python
# wap to find fibonacci number 

def fibo(n):
    a = 0
    b = 1
    
    if n==1 :
        return a 
    else:
        for i in range(1 , n+1):
            sum = a + b 
            a = b 
            b = sum 
        
        return sum

axy = fibo(4)
print(axy)
```

```python
# Class and Objects

class Cartesian:
    def __init__(self, xx, yy):
        self.x = xx
        self.y = yy

    def __str__(self):
        return f"The cartesian points are ({self.x}, {self.y})"

    def __repr__(self):
        return f"Nice one {self.x, self.y}"
    
    def __add__(self , other):
        new_x = self.x + other.x
        new_y = self.y + other.y 
        
        return Cartesian(new_x , new_y)

a1 = Cartesian(1, 4)
a2 = Cartesian(3 ,5)

a3 = a1+a2
print(a3)
```

```python
# home work WAP to create a class named Distance which
# has feet and inches as data member . Overload any two operators


class Distance:
    def __init__(self, feet, inches):
        if inches >= 12:
            quot = inches // 12
            rem = inches % 12

            self.feet = feet + quot
            self.inches = rem
        else:
            self.feet = feet
            self.inches = inches

    def __str__(self):
        return f"{self.feet} ft and {self.inches} inches"

    def __repr__(self):
        return f"{self.feet} , {self.inches}"

    def __add__(self, other):
        new_feet = self.feet + other.feet
        new_inches = self.inches + other.inches
        return Distance(new_feet, new_inches)

    def __mul__(self, other):
        if not isinstance(other, int):
            raise TypeError("Distance can only be multiplied by a number")

        new_feet = self.feet * other
        new_inches = self.inches * other
        return Distance(new_feet, new_inches)


# 1ft = 12 inches

d1 = Distance(3, 10)
d2 = Distance(4, 2)

d3 = d1 + d2
d4 = d1 * 4

print(f"The product is {d4} and sum is {d3}")
```
